## Duplicated features with Feature-engine

In this notebook, we will identify and remove duplicated features with Feature-engine.

```python
# import libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

from feature_engine.selection import DropConstantFeatures, DropDuplicateFeatures
```

```python
# load data

data = pd.read_csv('../precleaned-datasets/dataset_1.csv')
data.head()
```

```python
data.shape
```

**Important**

In all feature selection procedures, it is a good practice to select the features by examining only the training set. And this is to avoid overfit.

```python
# separate dataset into train and test

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['target'], axis=1), # drop the target
    data['target'], # just the target
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

## Remove constant and quasi-constant

```python
# remove constant and quasi-constant features first:
# we use Feature-engine for this

sel = DropConstantFeatures(tol=0.998, variables=None, missing_values='raise')

sel.fit(X_train)
```

```python
sel.features_to_drop_
```

```python
# remove the quasi-constant features

X_train = sel.transform(X_train)
X_test = sel.transform(X_test)

X_train.shape, X_test.shape
```

- Now we can see, we are left with only 158 features after removing the Constant and Quasi Constant features

## Remove duplicated features

```python
# set up the selector
sel = DropDuplicateFeatures(variables=None, missing_values='raise')

# find the duplicate features, this might take a while
sel.fit(X_train)
```

```python
sel.duplicated_feature_sets_
```

```python
sel.features_to_drop_
```

```python
# remove the duplicated features

X_train = sel.transform(X_train)
X_test = sel.transform(X_test)

X_train.shape, X_test.shape
```

- Now the duplicated features are also removed, so left with only 152 features, almost half of the original features set

## Stack Feature selection in a Pipeline

We can perform both steps together by setting up the transformers within a pipeline.

```python
# separate dataset into train and test

X_train, X_test, y_train, y_test = train_test_split(
    data.drop(labels=['target'], axis=1), # drop the target
    data['target'], # just the target
    test_size=0.3,
    random_state=0)

X_train.shape, X_test.shape
```

```python
pipe = Pipeline(steps = [
    ('constant', DropConstantFeatures(tol=0.998, variables=None, missing_values='raise')),
    ('duplicates', DropDuplicateFeatures(variables=None, missing_values='raise'))
])

pipe.fit(X_train)
```

```python
pipe.named_steps['duplicates'].features_to_drop_
```

```python
# remove features

X_train = pipe.transform(X_train)
X_test = pipe.transform(X_test)

X_train.shape, X_test.shape
```

```python
# we can navigate the pipeline transformers

len(pipe.named_steps['constant'].features_to_drop_)
```
